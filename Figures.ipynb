```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import numpy as np
```

```python
exp = pd.read_csv("Expression/ExpressionScores.csv",index_col=0)
sol = pd.read_csv("SWI/SolubilityScores.csv",index_col=0)
agg = pd.read_csv("Aggregation/AggregationScores.csv",index_col=0)
```

```python
exp
sns.set(style='ticks', font_scale=2)

title_size = 14
sat=1
plt.subplots(1, 1, figsize=(9,5), sharey=True)

sns.set_style(style = 'white')
sns.barplot(x ="metric", y = "score", data = exp,  ci = "sd", capsize=.1, palette ="Purples")

plt.xlabel("Performance measure", size=21)
plt.ylabel("Score", size=21)

plt.savefig('Performance_expression.png',dpi=300, bbox_inches='tight')
plt.show()
```

```python
sol
sns.set(style='ticks', font_scale=2)

title_size = 14
sat=1
plt.subplots(1, 1, figsize=(9,5), sharey=True)

sns.set_style(style = 'white')
sns.barplot(x ="metric", y = "score", data = sol,  ci = "sd", capsize=.1, palette ="Blues")

plt.xlabel("Performance measure", size=21)
plt.ylabel("Score", size=21)

plt.savefig('Performance_solubility.png',dpi=300, bbox_inches='tight')
plt.show()
```
